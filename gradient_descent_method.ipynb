```python
import numpy

def sigmoid(sop):
    return 1.0/(1+numpy.exp(-1*sop))  #1/1=e^-x                
def error(predicted,target):
    return numpy.power(predicted-target,2)#x^2;x=pred-tar
def error_predicted_deriv(predicted,target):
    return 2*(predicted-target)#d(x^2=2x)               
def activation_sop_deriv(sop):
    return sigmoid(sop)*(1.0-sigmoid(sop))#actual fxn
def sop_w_deriv(x):
    return x
def update_w(w,grad,learning_rate):
    return learning_rate *grad #updating weight
x=0.1
target=0.3
learning_rate=0.01     
w=numpy.random.rand()
print('w:', w)
for k in range(100):
    #forward pass            
    y=w+x
    predicted=sigmoid(y)
    err=error(predicted,target)
    #backward pass
    g1=error_predicted_deriv(predicted,target)
    g2=activation_sop_deriv(predicted)
    g3=sop_w_deriv(x)
    grad=g3*g2*g1
    print(predicted)
                
    w=update_w(w,grad,learning_rate)
    
```
